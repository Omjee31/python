```python
def decimal_to_binary(num):
    num1 = num
    bin_num = []
    if num == 0:
       return "0"
    while num>0:
        r = num%2
        bin_num.append(str(r))
        num = num//2
    return f"The Binary Of {num1} Is {''.join(reversed(bin_num))}"


# Example usage
print(decimal_to_binary(25)) # Output: 11001
print(decimal_to_binary(174)) # Output: 10101110
print(decimal_to_binary(31))  # output: 11111
```

```python
def decimal_to_octal(num):
    num1 = num
    octal_num = []
    if num == 0:
       return "0"
    while num>0:
        r = num%8
        octal_num.append(str(r))
        num = num//8
    return f"The Octal Of {num1} Is {''.join(reversed(octal_num))}"


# Example usage
print(decimal_to_octal(25))  # output: 31
print(decimal_to_octal(174)) # Output: 256
print(decimal_to_octal(31))  # output: 37

```

```python
def decimal_to_hex(num):
    num1 = num
    hex_num = []
    if num == 0:
       return "0"
    while num>0:
        r = num%16
        if r<10:
            hex_num.append(str(r))
        else:
            hex_num.append(chr(55+r))
        num = num//16
    return f"The Hex Number Of {num1} = {''.join(reversed(hex_num))}"


# Example usage
print(decimal_to_hex(25))  # output: 19
print(decimal_to_hex(174)) # Output: AE
print(decimal_to_hex(31))  # output: 1F
```

```python
def decimal_to_binary(num):
    if num == 0:
        return "0"
        
    digits = "0123456789ABCDEF"
    hex_num = ""
    
    while num > 0:
        hex_num = digits[num % 16] + hex_num
        num //= 16
        
    return hex_num


# Example usage
print(decimal_to_hex(25))  # output: 19
print(decimal_to_hex(174)) # Output: AE
print(decimal_to_hex(31))  # output: 1F
```

```python
def decimal_to_bin(num):
    if num == 0:
        return "0"
        
    digits = "01"
    bin_num = ""
    
    while num > 0:
        bin_num = digits[num % 2] + bin_num
        num //= 2
        
    return bin_num


# Example usage
print(decimal_to_binary(25)) # Output: 11001
print(decimal_to_binary(174)) # Output: 10101110
print(decimal_to_binary(31))  # output: 11111
```

```python
def decimal_to_octal(num):
    if num == 0:
        return "0"
        
    digits = "012345678"
    oct_num = ""
    
    while num > 0:
        oct_num = digits[num % 8] + oct_num
        num //= 8
        
    return oct_num

# Example usage
print(decimal_to_octal(25))  # output: 31
print(decimal_to_octal(174)) # Output: 256
print(decimal_to_octal(31))  # output: 37
```

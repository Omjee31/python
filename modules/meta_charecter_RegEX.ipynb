| Pattern | Meaning         |
| ------- | --------------- |
| `.`     | Any character   |
| `\.`    | Literal dot `.` |
. ^ $ * + ? { } [ ] \ | ( )  '\\' include special char

```python
import re
s = 'omjee singh is a good boy'
#without'\'
match = re.search(r'.',s)
print(match)
# with'\'
match = re.search(r'\.',s)
print(match)
```

### [] is  used where inside we keep our pattern
##### ex :- [a-m],[0-3]

```python
pattern = '[a-o]'
match = re.findall(pattern,s)
print(match)
```

### ^ - caret  use to cheak begining

```python
s = 'The OMjee singh'
pattern = r'^The'
for S in s:
    if re.match(pattern,S):
        print(f'matched :{S}')

    else:
        print(f'Not Matched :{S}')
```

### $ - dollar cheak end of string

```python
s = 'The OMjee singh'
pattern = r'singh$'
match = re.search(pattern,s)
if match:
    print(f'matched found')

else:
    print(f'Not Matched ')

print(match)
```

#### '\b' uuse to cheak given pattern

```python
result = re.search(r'\bOMjee\b',s)
print(result)
print(result.start())
print(result.end())
print(result.span())
```

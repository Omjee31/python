Find the longest word in a sentence.

```python
from collections import Counter
import re  #with RegEx

# Open file and read content
with open("data.txt", "r") as file:
    text = file.read().lower()   # Convert to lowercase

# Remove punctuation
text = re.sub(r"[^\w\s]", "", text)

# Split into words
words = text.split()


# ---- by using dict comprehension
freq = {word:len(word) for word in words}
longest_word = max(freq, key=freq.get)
print(longest_word)
```

```python
# simple and better way
longest = max(words,key=len)
print(longest)
```

```python
longest_word = ""

for word in words:
    if len(word) > len(longest_word):
        longest_word = word

print(longest_word)
```

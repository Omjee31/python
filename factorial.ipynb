```python
def factorial_Recursion(n):
    """Compute factorial using recursion."""
    if n == 0 or n == 1:
        return 1
    return n * factorial_Recursion(n - 1)

def fact(n):
    """Compute factorial using iterative loop."""
    factorial = 1
    while n > 1:  # Minor optimization: stop at 1 instead of 0
        factorial *= n
        n -= 1
    return factorial

# Interactive demo
if __name__ == "__main__":
    num = int(input("Enter The Number You Want That Factorial: "))
    print(f"Factorial Of {num} Using Recursion: {factorial_Recursion(num)}")
    print(f"Factorial Of {num} Using Loop: {fact(num)}")
```

 Find the second largest unique element in a list.

```python
def unique_second_largest(list1):
    unique_numbers = sorted(set(list1))  # Remove duplicates + sort
    
    if len(unique_numbers) < 2:
        return "Not enough unique elements"
    
    return f'Second Highest Unique Element is {unique_numbers[-2]}'


list1 = [12,34,1,2,56,78,45,32,48,99,78,98]
print(unique_second_largest(list1))
```

```python
def unique_second_largest(list1):
    first = second = float('-inf')
    
    for num in set(list1):  # Remove duplicates
        if num > first:
            second = first
            first = num
        elif num > second:
            second = num
            
    if second == float('-inf'):
        return "Not enough unique elements"
    
    return f'Second Highest Unique Element is {second}'

list1 = [12,34,1,2,56,78,45,32,48,99,78,98]
print(unique_second_largest(list1))
```

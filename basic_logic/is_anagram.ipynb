## What is anagrams
> In Python, anagrams are words or phrases that contain the exact same characters with the same frequencies, regardless of the character order. For example, "listen" and "silent" are anagrams, as are "heart" and "earth".

```python
def is_anagrams_sorted(str1,str2):
    str1 = str1.lower()
    str2=str2.lower()

    if len(str1)!=len(str2):
        return False

    return sorted(str1)==sorted(str2)


# Example usage
print(is_anagram_sorted("listen", "silent")) # Output: True
print(is_anagram_sorted("Race", "Care"))   # Output: True
print(is_anagram_sorted("hello", "world")) # Output: False
```

```python
from collections import Counter
def is_anagrams_using_counter(str1,str2):
    str1 = str1.lower()
    str2=str2.lower()

    if len(str1)!=len(str2):
        return False

    return Counter(str1)==Counter(str2)


# Example usage
print(is_anagram_sorted("listen", "silent")) # Output: True
print(is_anagram_sorted("Race", "Care"))   # Output: True
print(is_anagram_sorted("hello", "world")) # Output: False
```

4 Rotate a list by k positions (left & right).



```python
def rotate_list_right(nums, k):
    n = len(nums)
    if n == 0:
        return nums
    # Calculate the effective number of rotations
    k %= n
    # Slice the list and concatenate in the new order
    rotated_list = nums[n-k:] + nums[:n-k]
    return rotated_list

# Example usage
my_list = [1, 2, 3, 4, 5]
k_places = 2
rotated = rotate_list_right(my_list, k_places)
print(f"Original list: {my_list}")
print(f"Rotated list (by {k_places} places): {rotated}")
```

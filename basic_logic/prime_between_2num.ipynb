```python
#Write a function that returns all prime numbers between two given numbers.
import time
def check_prime(num1, num2):
    start = time.time()
    primes = []
    
    for num in range(num1, num2 + 1):
        
        if num < 2:
            continue
        
        is_prime = True
        
        # check divisibility up to sqrt(num)
        for i in range(2, int(num**0.5) + 1):
            if num % i == 0:
                is_prime = False
                break
        
        if is_prime:
            primes.append(num)
    
   # print(primes)
    end = time.time()
    return f"Time During This Is {start-end}"


print(check_prime(10, 500000))
```

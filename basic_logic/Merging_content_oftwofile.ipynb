Merge contents of two files into a third file.

```python
# copy one binary file into another 
with open('ai3.png', 'rb') as f, open('Ai.png', 'wb') as wr:
    wr.write(f.read())
```

```python
# add Content of two txt file into third txt file
with open('untitled.txt', 'r') as f, open('untitled1.txt', 'r') as r, open('New.txt','w')as w:
    d1 = f.read()
    d2 = r.read()
    d3 = d1+d2
    w.write(d3)
    
```
